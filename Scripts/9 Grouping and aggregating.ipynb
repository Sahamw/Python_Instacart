```python
## 6 The team now wants to target different types of spenders in their marketing campaigns. 
##If the mean of the prices of products purchased by a user is lower than 10, then flag them as a Low spender.
## If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a High spender.

ords_prods_merge['mean_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_spend'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_spend'] >= 10, 'spending_flag'] = 'High spender'
```

```python
## 7 Create an order frequency flag that marks the regularity of a users ordering behavior according to 
##the median in the days_since_prior_order column

#If the median of days_since_prior_order is higher than 20, then the customer should be labeled a Non-frequent customer.
#If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a Regular customer.
#If the median is lower than or equal to 10, then the customer should be labeled a Frequent customer.
```

```python
ords_prods_merge['median_order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order_frequency'] <= 20) & (ords_prods_merge['median_order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```
